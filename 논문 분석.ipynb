# 형태소 분석

```python
# pip install kiwipiepy

from kiwipiepy import Kiwi
kiwi = Kiwi()
```

```python
kiwi.tokenize(word_data_list)
```

```python
# 명사 추출 함수
def extract_noun(text):
    result = kiwi.tokenize(text)
    for token in result:
        if token.len > 1 and token.tag in ['NNG', 'NNP']: # 은/는/이/가와 같은 조사(1글자)는 필요없으므로 버린다. 하지만 and 대신 or 을 사용하므로써 명사는 살려놓을 수 있다.
          yield token.form
```

```python
data_noun_list = []
for data in word_data_list:
  data_noun_list.append(list(extract_noun(data)))
```

```python
data_noun_list[:5]
```

# DTM

```python
from sklearn.feature_extraction.text import CountVectorizer
```

max_features: 단어문서행렬에 포함시킬 최대(max)의 단어(feature) 수

tokenizer: 정의한 형태소 분석기 설정

```python
cv = CountVectorizer(max_features=1000,tokenizer=extract_noun)
```

```python
tdm = cv.fit_transform(word_data_list) #term document metrix
```

```python
tdm
```

```python
tdm.sum(axis=0)  # 단어 빈도
```

```python
list(tdm.sum(axis=0).flat)
```

```python
len(list(tdm.sum(axis=0).flat))
```

```python
cv.get_feature_names_out()
```

```python
import pandas as pd
dtm_df = pd.DataFrame({
    '단어': cv.get_feature_names_out(),
    '빈도': tdm.sum(axis=0).flat
})
```

```python
dtm_df.head()
```

```python
dtm_df.sort_values('빈도', ascending=False).head(20)
```

```python
# pip uninstall numpy scipy gensim networkx tensorflow matplotlib
# pip install numpy==1.24.0 scipy==1.10.0 gensim==4.3.1 networkx==3.0 tensorflow==2.14.0 matplotlib==3.6

from gensim.models import Word2Vec
```

 - window : 문장내에서 현재 단어와 예측 단어 사이의 최대 거리
 - vector_size : 단어 임베딩 차원
 - sg : 1=skip-gram, 0=CBOW
 - min_count : 총 빈도가 낮은 단어를 무시하는 옵션

```python
w2v_model = Word2Vec(data_noun_list, window=5, vector_size=100, sg=1, min_count=5)
```

```python
w2v_model.wv.vectors.shape
```

```python
w2v_model.wv.most_similar('모델', topn=20) # '모델'과 연관성이 가장 높은 단어 20개
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
import networkx as nx
import os

# 폰트 경로
font_path = 'C:/Users/kimsu/Desktop/Crawling/NotoSansKR-VariableFont_wght.ttf'
if not os.path.exists(font_path):
    raise FileNotFoundError(f"Font file not found: {font_path}")

# 폰트 등록
font_manager.fontManager.addfont(font_path)
font_name = font_manager.FontProperties(fname=font_path).get_name()

# 폰트 설정
font_prop = font_manager.FontProperties(fname=font_path)
font_manager.fontManager.addfont(font_path)
rc('font', family=font_prop.get_name())

plt.rcParams['font.family'] = font_name  
plt.rcParams['axes.unicode_minus'] = False

# 만약 sans-serif 계열로도 지정하고 싶다면 추가
plt.rcParams['font.sans-serif'] = [font_name]

# 혹은 rc()로 지정해도 됨
# rc('font', family=font_name)
# rc('axes', unicode_minus=False)

def create_language_network(model, topn=10):
    G = nx.Graph()
    words = list(model.wv.index_to_key)
    for word in words:
        try:
            similar_words = model.wv.most_similar(word, topn=topn)
            for sw, sim in similar_words:
                G.add_edge(word, sw, weight=sim)
        except KeyError:
            continue
    return G

# 네트워크 생성
G = create_language_network(w2v_model, topn=5)

plt.figure(figsize=(18, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='black', alpha=0.01)
nx.draw_networkx_edges(G, pos, width=0.3, alpha=0.35)
nx.draw_networkx_labels(G, pos, font_size=11, font_color='b')
plt.title(f'Word2Vec Language Network word count - {len(G.nodes)}', fontsize=15)
plt.show()
```
